# DATA CLEANING

```python
filename01 = "credit_card_balance.csv"
filepath = ROOT_DIR + "data/" + filename01
# data savings
if not os.path.exists(ROOT_DIR + "data/"):
    os.mkdir(ROOT_DIR + "data/")
# data saving
if not os.path.exists(ROOT_DIR + "figures/"):
    os.mkdir(ROOT_DIR + "figures")

# Reading of the raw data
df_raw_data01 = pd.read_csv(filepath, sep=",",low_memory=False, header=0, encoding='utf-8')

df_raw_data01.head(3) 
```

```python
#remplacement des "faux NaN" en NaN reconnus par pandas
df_raw_data01.replace(to_replace = '^nan$', value = np.nan, regex=True,inplace=True)
df_raw_data01.describe(include="all")
```

```python
df_raw_data01.dtypes
```

### Analyse du taux de données manquantes

```python

proportion_nan_raw01 = df_raw_data01.isna().sum().divide(df_raw_data01.shape[0]/100).sort_values(ascending=False)
```

```python
# Visualisation graphique du taux de remplissage  
sns.set(style="whitegrid")
plt.figure(figsize=(16, 30))
ax = sns.barplot(y = proportion_nan_raw01.index, x=proportion_nan_raw01.values)
ax.xaxis.set_ticks_position('top')
plt.title('Quantité de données manquantes par colonne dans le jeu de données (en %)', size=15)
plt.savefig(ROOT_DIR + 'figures/taux_missing_data_all.png')
plt.show()
```

```python
# Visualisation graphique du taux de remplissage  
count_null01 = {'variables': proportion_nan_raw01.index,
              'percentage of null values':proportion_nan_raw01.values}
df_count_null01 = pd.DataFrame(data=count_null01)

fig_data_usage = px.bar(df_count_null01, y='variables', x='percentage of null values',
                        title='Quantité de données manquantes par colonne dans le jeu de données (en %)')
fig_data_usage.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 10
    ),
    yaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 1
    )
)
```
